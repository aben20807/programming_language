```python
import numpy as np
plt.rcParams['font.size']=10
x = np.array([l1,l2,l3,l4,l5,l6,l7,l8,l9,l10,l11,l12]).T
print(x)
df = pd.DataFrame(x,  columns=list('ABCDEFGHIJKL'))
df = df.cumsum()
result = df.plot(title = 'parallel matrix computation',ylim=[0,600])
result.set_xlabel("n*n")
result.set_ylabel("ms")
#result.set_xticklabels(['10','50','100','150','200','250','300','350'])

plt.xticks([0,5,10,15,20,25,30],[r'$10$', r'$50$', r'$100$', r'$150$', r'$200$', r'$250$',r'$300$'])

fig = result.get_figure()
fig.set_size_inches(8, 5)
fig.savefig("line_all.png")
```
